# Medical Device Data – Exploratory Analysis

This notebook demonstrates how to load, preprocess, and visualize medical device time-series data.

```python
import sys
sys.path.insert(0, '../src')

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from meddevice.data.preprocessing import normalize, remove_baseline_wander
```

## 1. Simulate an ECG-like signal

```python
# Synthetic signal: 10 seconds @ 250 Hz with a slow baseline wander
fs = 250  # Hz
t = np.linspace(0, 10, fs * 10)

# ECG-like component
ecg_like = np.sin(2 * np.pi * 1.2 * t)  # ~72 bpm

# Baseline wander (very low frequency)
wander = 0.5 * np.sin(2 * np.pi * 0.05 * t)

raw_signal = ecg_like + wander
```

## 2. Preprocess: remove baseline wander and normalize

```python
clean_signal = remove_baseline_wander(raw_signal, window=int(fs * 2))
norm_signal  = normalize(clean_signal)
```

## 3. Visualize

```python
fig, axes = plt.subplots(3, 1, figsize=(12, 8), sharex=True)

axes[0].plot(t, raw_signal, color='steelblue')
axes[0].set_title('Raw signal (with baseline wander)')
axes[0].set_ylabel('Amplitude')

axes[1].plot(t, clean_signal, color='darkorange')
axes[1].set_title('After baseline removal')
axes[1].set_ylabel('Amplitude')

axes[2].plot(t, norm_signal, color='forestgreen')
axes[2].set_title('Normalized [0, 1]')
axes[2].set_ylabel('Amplitude')
axes[2].set_xlabel('Time (s)')

plt.tight_layout()
plt.show()
```
